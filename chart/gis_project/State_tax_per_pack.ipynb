```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/nadezdachikurova/Documents/QGIS/Final project/data/TAX/State_tax_per_pack.csv'
                 , encoding='latin-1')
df1.head()
```

```python
df1["GeoLocation"].head()
```

```python
df1.dropna(inplace = True) 
df1.head()
```

```python
df1["GeoLocation"].head()
```

```python
df1[["Latitude_1","Longitude_1"]] = df1.GeoLocation.str.split(",",expand=True)

df1.head()
```

```python
df1[["extra","Latitude"]] = df1.Latitude_1.str.split("(",expand=True)
df1.head()
```

```python
df1[["Longitude","drop"]] = df1.Longitude_1.str.split(")",expand=True)
df1.head()
```

```python
df2 = df1.drop(columns = ["Longitude_1","Latitude_1","extra","drop"])
df2.head()
```

```python

df2["Id"] = df2.index + 1
df2.head()
```

```python
df2.to_csv(r'/Users/nadezdachikurova/Documents/QGIS/Final project/data/TAX/State_tax_per_pack_L.csv')
```
