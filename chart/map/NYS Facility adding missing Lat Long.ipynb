```python
from geopy.extra.rate_limiter import RateLimiter

#Creating a dataframe with address of locations we want to reterive
locat = ['116 Martin Luther King East','14 DeKalb Avenue','1808 Crotona Parkway','175 Tarrytown Road','120 Church Street']
df5 = pd.DataFrame({'add': locat})
 
#Creating an instance of Nominatim Class
geolocator = Nominatim(user_agent="my_request")
 
#applying the rate limiter wrapper
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
 
#Applying the method to pandas DataFrame
df5['location'] = df5['add'].apply(geocode)
df5['Lat'] = df5['location'].apply(lambda x: x.latitude if x else None)
df5['Lon'] = df5['location'].apply(lambda x: x.longitude if x else None)
 
df5
```
