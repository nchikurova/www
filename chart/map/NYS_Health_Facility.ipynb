```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/nadezdachikurova/Documents/School MS/QGIS/Lab_Assignment_01_Data/NYS_Retail_Food_Stores.csv'
                 , encoding='latin-1')
df1.head()
```

```python
df1["Location"].head()
```

```python
df1.dropna(inplace = True) 
df1.head()
```

```python
df1["Location"].head()
```

```python
df1[["Address","Coords","none"]] = df1.Location.str.split("(",expand=True)

df1.head()
```

```python
df1[["Latitude","Longitude)"]] = df1.Coords.str.split(",",expand=True)
df1.head()
```

```python
df2 = df1.drop(columns = ["Location","none","Coords"])
df2.head()
```

```python
df2[["Longitude","new"]] = df2['Longitude)'].str.split(")",expand=True)
df2.head()
```

```python
df3 = df2.drop(columns = ["Longitude)","new"])
df3.head()
```

```python
df3.to_csv(r'/Users/nadezdachikurova/Documents/School MS/QGIS/Lab_Assignment_01_Data/NYS_Retail_Food_Stores_NEW.csv')
```
